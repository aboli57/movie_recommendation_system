# Preprocessing

```python
import ast
ast.literal_eval('[{"name":"Adventure"},{"Fantasy"},{"id": 28, "name": "Action"}]')
```

```python
def convert(text):
    ext = []
    for i in ast.literal_eval(text):
        ext.append(i["name"])
    return ext
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
def convert1(text):
    ext1 = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 5:
            ext1.append(i["name"])
        counter +=1
    return ext1
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
def fetch_director(text):
    ext2 = []
    for i in ast.literal_eval(text):
        if i['job']=="Director":
            ext2.append(i['name'])
    return ext2
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(ext):
    ext3 = []
    for i in ext:
        ext3.append(i.replace(" ",""))
    return ext3
```

```python
movies['crew'] = movies['crew'].apply(collapse)
```

```python
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

# Feature Engineering Methods - Bag of Words (BOW) OR Term Frequency - Inverse Document (TF-IDF Vectorizer)

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(new_data['context']).toarray()
vector.shape
```

```python
pd.DataFrame(vector).head()
```

# Building Recommendation System by using Cosine-Similarity approach

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
import pickle
pickle.dump(new_data, open('movies.pkl','wb'))
pickle.dump(similarity, open('similarity.pkl','wb'))
```

```python
def recommendationsystem(movies):
    index = new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key= lambda x: x[1])
    for i in distance[1:6]:
        print(new_data.iloc[i[0]].title)
```
